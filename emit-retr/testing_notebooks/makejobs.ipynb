```python
def get_header_string(loc_name, prefix):
    return f"""#!/bin/bash
#SBATCH --job-name={prefix}_{loc_name}
#SBATCH --output=jobs/out/{prefix}_{loc_name}_%j.out
#SBATCH --error=jobs/out/{prefix}_{loc_name}_%j.err
#SBATCH --time=12:00:00
#SBATCH --ntasks=1
#SBATCH --cpus-per-task=2
#SBATCH --mem=64G

module load miniforge
conda activate nox_env_2

cd /home/kgauld/emit-nox-plumes/emit-retr"""

# python plot_retrieval_fields.py --loc_name {loc_name} --loc_tick --wind_tick --tv


with open("../targets1", "r") as f:
    target_names = f.read().splitlines()
target_names
```

```python
prefix = 'tavg'
for loc_name in target_names:
    hstring = get_header_string(loc_name, prefix)
    hstring += f"\npython BATCH_RETR.py --loc_name {loc_name} --time_avg --no_rerun"
    #f"\npython BATCH_RETR.py --loc_name {loc_name} --skip_done --include_mask --km_boundary 40"

    with open(f"../../../jobs/{prefix}_{loc_name}.sh", 'w') as f:
        f.write(hstring)
```

```python
import subprocess

for loc_name in target_names:
    if loc_name == 'Intermountain':
        print("skipping intermountain")
        continue
    result = subprocess.run(
        ["sbatch", f"../../../jobs/{prefix}_{loc_name}.sh"],
        capture_output=True,
        text=True,
        check=True
    )
    
    print(result.stdout)
```
